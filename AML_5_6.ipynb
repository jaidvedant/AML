```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/HousingData.csv'

# Robust Data Loading
try:

    data = pd.read_csv(file_path)
    if len(data.columns) < 10: raise ValueError
except:
    # Fallback for whitespace-delimited files
    column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
    data = pd.read_csv(file_path, header=None, sep=r'\s+', names=column_names)

# Convert all to numeric (handles headers read as data)
data = data.apply(pd.to_numeric, errors='coerce').dropna()

# EDA: Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', fmt='.2f')
plt.title("Pune Housing: Feature Correlation Analysis")
plt.show()
```
